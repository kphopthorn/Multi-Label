```python
import fasttext
```

```python
pretrain_path = '/content/drive/MyDrive/IS_multi-label/cc.th.300.vec'
train_path = '/content/drive/MyDrive/IS_multi-label/train_fasttext.txt'
```

```python
#model = fasttext.train_supervised(**fasttext_params)
model = fasttext.train_supervised(train_path,epoch=50, dim=300, wordNgrams = 2,loss='ova',lr=0.5, bucket=200000,verbose=1)
                                  #pretrainedVectors=pretrain_path)
print('vocab size: ', len(model.words))
print('label size: ', len(model.labels))
print('example vocab: ', model.words)
print('example label: ', model.labels)
```

```python
model.predict('ไฟแดง เลี้ยว ขวา เข้า ซอย เสีย ทำ ให้ รถ เลี้ยว ฝ่า ไฟ แดง ใน ตอน ทางตรง ไฟ เขียว อันตราย มาก ครับ', k=24,threshold=0.5)
```

```python
num_param = (len(model.words) * model.dim) + (model.dim *len(model.labels))
print(f'number of Param : {num_param}')
```

```python
import pandas as pd
cols = ['PM2.5', 'การเดินทาง', 'กีดขวาง', 'คนจรจัด', 'คลอง',
        'ความปลอดภัย', 'ความสะอาด', 'จราจร', 'ต้นไม้', 'ถนน', 'ทางเท้า',
        'ท่อระบายน้ำ', 'น้ำท่วม', 'ป้าย', 'ป้ายจราจร', 'ร้องเรียน', 'สอบถาม',
        'สะพาน', 'สัตว์จรจัด', 'สายไฟ', 'ห้องน้ำ', 'เสนอแนะ', 'เสียงรบกวน',
        'แสงสว่าง']
output_df = pd.DataFrame(columns=cols)
```

```python
df_test = pd.read_csv('/content/drive/MyDrive/IS_multi-label/test_data.csv')
```

```python
for j in df_test['comm_norm']:
    x = model.predict(j, k=24)
    new_row = {label.replace('__label__', ''): value for label, value in zip(x[0], x[1])}
    output_df = output_df.append(new_row, ignore_index=True)
```

```python
eval = model.test("/content/drive/MyDrive/IS_multi-label/test_fasttext.txt")
```

```python
sample,precision,recall = eval
print(f'{sample} samples, precision = {precision}, recall = {recall}')
print(f'accuracy = {2 *( (precision *recall) / (precision + recall))}')
```

```python
import tensorflow_addons as tfa
```

```python
y_true = df_test[cols]
```

```python
hl_metric = tfa.metrics.HammingLoss(mode='multilabel', threshold=0.5)
hl_metric.update_state(y_true, output_df)
```

```python
hl_metric.result().numpy()
```

```python
y_pred_df = output_df.applymap(lambda x: 1 if x >= 0.5 else 0)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_true, y_pred_df, target_names=cols))
```
