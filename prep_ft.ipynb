```python
ft_cols = data.columns[4:]
ft_cols
```

```python
selected_columns = []

# Iterate through the DataFrame to check values and collect column names
for index, row in data.iterrows():
    selected_column_names = [f"__label__{col}" for col in ft_cols if row[col] == 1]
    selected_columns.append(" ".join(selected_column_names))

# Create a new column in the DataFrame with the concatenated values
data['new_col_name'] = selected_columns
```

```python
 data['new_col_name']
```

```python
data['fasttext_train'] =  data['new_col_name'] + " " + data['comm_norm']
```
